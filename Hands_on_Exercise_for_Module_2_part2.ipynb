```python
from rdkit import Chem
from rdkit.Chem import Draw

# Define molecules
molecules = [
    ("Trans-2-butene", "C/C=C/C"),
    ("Cis-2-butene", "C/C=C\\C"),
    ("(R)-2-Butanol", "C[C@H](O)CC"),
    ("(S)-2-Butanol", "C[C@@H](O)CC"),
    ("Cyclopentane", "C1CCCC1"),
    ("Cyclopropane", "C1CC1"),
    ("2-Methylpentane", "CC(C)CCC"),
    ("Butanol", "CCCCO")
]

mols = []
legends = []

# Build lists safely
for name, smi in molecules:
    mol = Chem.MolFromSmiles(smi)
    if mol is not None:
        mols.append(mol)
        legends.append(name)
    else:
        print(f"❌ Invalid SMILES: {name}")

# 🔴 Critical check
print("Molecules:", len(mols))
print("Legends:", len(legends))

# Ensure equal length
assert len(mols) == len(legends), "Mismatch between mols and legends!"

# Draw
img = Draw.MolsToGridImage(
    mols,
    molsPerRow=4,
    subImgSize=(200, 200),  # <-- note correct parameter name
    legends=legends
)

img
```

```python
!pip install biopython

from Bio.Seq import Seq
from Bio.Blast import NCBIWWW, NCBIXML

# 1. Translation
# Note: Ensure your DNA sequence is a valid coding sequence for A. baumannii
dna = Seq("ATGGCCATTGTAATGGGCCGCTGA")
protein_seq = dna.translate(to_stop=True)
print(f"Translated Protein: {protein_seq}")

# 2. Targeted BLAST Search
print("Searching specifically within Acinetobacter baumannii records...")

# 'entrez_query' acts as a filter.
# "txid470[ORGN]" is the specific Taxonomy ID for Acinetobacter baumannii
result_handle = NCBIWWW.qblast(
    "blastp",
    "nr",
    protein_seq,
    entrez_query="Acinetobacter baumannii[ORGN]"
)

# 3. Parsing the Results
blast_record = NCBIXML.read(result_handle)

print("\n--- A. baumannii BLAST Results ---")
if blast_record.alignments:
    for alignment in blast_record.alignments:
        print(f"Match: {alignment.title}")
        e_value = alignment.hsps[0].expect
        print(f"E-value: {e_value}")

        # Highlighting Identity (How much of the sequence matches exactly)
        identity = alignment.hsps[0].identities
        align_length = alignment.hsps[0].align_length
        print(f"Identity: {identity}/{align_length} ({(identity/align_length)*100:.2f}%)")
        break
else:
    print("No matches found within Acinetobacter baumannii. This sequence might not exist in this species.")
```

```python
# Step 1: Setup and Data Import
# Install RDKit in your Colab environment
!pip install rdkit
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Simple Dataset: 5 Molecules (SMILES) and their solubility (logS)
data = {
'Compound': ['Methane', 'Ethane', 'Propane', 'Butane', 'Pentane'],
'LogP': [1.09, 1.81, 2.36, 2.89, 3.39], # Predictor (Input)
'Actual_LogS': [-1.5, -2.2, -3.0, -3.8, -4.5] # Actual Value (Output)
}
df = pd.DataFrame(data)
print(df)
```

```python
# Step 2: Perform Linear Regression
# Here, we create our "Brain" (the model) and show it the data.
# Prepare the data
X = df[['LogP']] # The "Question"
y = df['Actual_LogS'] # The "Answer"
# Initialize and Train the Model
model = LinearRegression()
model.fit(X, y)
# Make Predictions
df['Predicted_LogS'] = model.predict(X)
print(df[['Compound', 'Actual_LogS', 'Predicted_LogS']])
```

```python
# Step 3: Visualize Results
# We will draw a plot to see how close the computer's "guesses" were to the truth.
plt.scatter(df['LogP'], df['Actual_LogS'], color='blue', label='Actual Data')
plt.plot(df['LogP'], df['Predicted_LogS'], color='red', label='ML Prediction Line')
plt.xlabel('LogP (Lipophilicity)')
plt.ylabel('LogS (Solubility)')
plt.title('ML Prediction of Solubility')
plt.legend()
plt.show()
```

```python
import numpy as np
# 1. Define the new LogP value
# We put it in a 2D array [[5.0]] because the model expects a list of samples
new_logp = np.array([[5.0]])
# 2. Use the trained model to predict the LogS
predicted_logs = model.predict(new_logp)
# 3. Print the result
print(f"For a compound with LogP = 5.0, the predicted LogS is: {predicted_logs[0]:.2f}")
```
